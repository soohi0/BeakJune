```python
#BFS
import copy
from collections import deque
n, m = map(int, input().split(' '))
D = [[0, 1], [0, -1],[-1, 0], [1, 0] ]
graph = []
st = []
town = 0
def isRange(x, y):
    if x >= 0 and x < m and y >= 0 and y < n:
        return True
    return False
for _ in range(n):
    string = input()
    li = []
    for idx, s in enumerate(string):
        li.append(int(s))
        if int(s) == 0:
            st.append([idx, _]) #x, y
        elif int(s) == 1:
            town += 1
    graph.append(li)

mintime = 400


for idx, sp in enumerate(st[:len(st)-1]):
    for sp2 in st[idx+1:]:
        print(sp2)
        print(sp)
        q = deque([])
        testgraph = copy.deepcopy(graph)
        q.append(sp)
        q.append(sp2)
        towncnt = 0
        time = 0
        while q:
            #not empty
            front = q.popleft()
            isthereSecond = False
            if q:
                second = q.popleft()
                isthereSecond = True
            if testgraph[front[1]][front[0]] == 1:
                towncnt += 1
            if isthereSecond:
                if testgraph[second[1]][second[0]] == 1:
                    towncnt += 1
            testgraph[front[1]][front[0]] = -1
            if isthereSecond:
                testgraph[second[1]][second[0]] = -1
            time += 1
            for dir in D:
                nextX = front[0] + dir[0]
                nextY = front[1] + dir[1]
                if isRange(nextX, nextY):
                    if testgraph[nextY][nextX] == 0 or testgraph[nextY][nextX] == 1:
                        q.append([nextX, nextY])
                        # testgraph[nextY][nextX] = -1
                if isthereSecond:
                    nextSX = second[0] + dir[0]
                    nextSY = second[1] + dir[0]
                    if isRange(nextSX, nextSY):
                        if testgraph[nextSY][nextSX] == 0 or testgraph[nextSY][nextSX] == 1:
                            q.append([nextSX, nextSY])
            if towncnt == town:
                break
                
        if time < mintime:
            mintime = time
print(mintime)
```
